```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgbm

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV 

from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv')
```

```python
df.head()
```

```python
df = df.drop('stab', axis=1)

X = df.drop('stabf', axis = 1)
y = df['stabf']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
scaler = StandardScaler()
```

```python
x_train_trans = scaler.fit_transform(x_train)
x_train_trans = pd.DataFrame(x_train_trans, columns = x_train.columns)
x_test_trans = scaler.transform(x_test)
x_test_trans = pd.DataFrame(x_test_trans, columns = x_test.columns)
```

```python
# What is the accuracy on the test set using the random forest classifier? In 4 decimal places.

rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(x_train_trans, y_train)
```

```python
preds = rf_model.predict(x_test_trans)
```

```python
accuracy = accuracy_score(y_test, preds)
```

```python
round(accuracy,4)
```

```python
# What is the accuracy on the test set using the XGboost classifier? In 4 decimal places
xgboost_model = XGBClassifier(random_state = 1)
xgboost_model.fit(x_train_trans, y_train)
```

```python
xgboost_pred = xgboost_model.predict(x_test_trans)
```

```python
xgboost_accuracy = accuracy_score(y_test,xgboost_pred )
```

```python
round(xgboost_accuracy,4)
```

```python
# What is the accuracy on the test set using the LGBM classifier? In 4 decimal places

lgbm_model= lgbm.LGBMClassifier(random_state=1)
lgbm_model.fit(x_train_trans, y_train)
lgbm_prediction  = lgbm_model.predict(x_test_trans)
round(accuracy_score(y_test, lgbm_prediction),4)
```
